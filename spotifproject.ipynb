```python
alpha = 0.001
num_iterations = 100000
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def gradientDescent(x, y, alpha, num_iterations):
    m, n = x.shape
    theta = np.zeros((n, 1))

    for iteration in range(num_iterations):
        z = np.dot(x, theta)
        h = sigmoid(z)
        gradient = np.dot(x.T, (h - y.reshape(-1, 1))) / m  # reshape y for proper broadcasting
        theta -= alpha * gradient

        if iteration % 100 == 0:
            print(f'Iteration {iteration}: Theta = {theta.T}, y_hat = {h.T}')

    return theta
```

```python
def naiveBayes(train_features, train_labels, test_sample):
    # Calculate class probabilities
    class_counts = {}
    for label in train_labels:
        class_counts[label] = class_counts.get(label, 0) + 1

    class_probs = {}
    total_samples = len(train_labels)
    for label, count in class_counts.items():
        class_probs[label] = count / total_samples

    # Calculate feature probabilities
    feature_probs = {}
    for label in class_probs:
        label_indices = [i for i in range(len(train_labels)) if train_labels[i] == label]
        label_features = [train_features[i] for i in label_indices]
        feature_probs[label] = []
        for i in range(len(label_features[0])):
            feature_column = [row[i] for row in label_features]
            feature_values = list(set(feature_column))
            prob_values = []
            for value in feature_values:
                prob = feature_column.count(value) / len(label_indices)
                prob_values.append(prob)
            feature_probs[label].append((feature_values, prob_values))

    # Calculate probabilities for test sample
    max_prob = 0
    predicted_label = None
    for label, class_prob in class_probs.items():
        feature_probabilities = feature_probs[label]
        prob = class_prob
        for i in range(len(test_sample)):
            feature_value = test_sample[i]
            feature_value_probs = feature_probabilities[i][1]
            if feature_value in feature_probabilities[i][0]:
                prob *= feature_value_probs[feature_probabilities[i][0].index(feature_value)]
            else:
                prob *= 1
        if prob > max_prob:
            max_prob = prob
            predicted_label = label

    return predicted_label, max_prob
```

```python
import math
from collections import Counter
```

```python
def kNN(training_data, training_label, k, test_sample):
    results = []

    for i in range(len(training_data)):
        distance = math.sqrt(sum((float(a) - float(b)) ** 2 for a, b in zip(training_data[i], test_sample)))
        results.append((distance, training_label[i]))

    results.sort(key=lambda x: x[0])
    k_nearest_label = [label for (_, label) in results[:k]]

    label_count = Counter(k_nearest_label)
    most_common_label, count = label_count.most_common(1)[0]

    if list(label_count.values()).count(count) > 1:
        return 'UNIDENTIFIED'
    else:
        return most_common_label
```

```python
print("For test data 1")
for k in range(1, 200):
    result = kNN(train_data, train_labels, k, test_data)
    print(f'k = {k}: {result}')

print("For test data 2")
for k in range(1, 200):
    result = kNN(train_data, train_labels, k, test_data4)
    print(f'k = {k}: {result}')

print("For test data 3")
for k in range(1, 200):
    result = kNN(train_data, train_labels, k, test_data5)
    print(f'k = {k}: {result}')
```
